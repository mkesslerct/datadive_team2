```python
from dbfread import DBF
import zipfile
from fastkml import kml
from pyproj import CRS
import geopandas as gpd

# Leer archivo .dbf
def read_dbf_file(file_path):
    table = DBF(file_path)
    data = []
    for record in table:
        data.append(record)
    return pd.DataFrame(data)

# Leer archivo .kmz
def read_kmz_file(file_path):
    with zipfile.ZipFile(file_path, 'r') as zip_ref:
        zip_ref.extractall('/path/to/extract/to')
    with open('/path/to/extract/to/doc.kml', 'rt', encoding='utf-8') as kml_file:
        doc = kml_file.read()
    k = kml.KML()
    k.from_string(doc)
    return k

# Leer archivo .prj
def read_prj_file(file_path):
    return CRS.from_file(file_path)

# Leer archivos .shp y .shx
def read_shp_file(file_path):
    return gpd.read_file(file_path)
```

```python
tabla_evolucion = read_dbf_file(DATA_DIRECTORY / 'SourceDataFig2.dbf')
tabla_evolucion
```

```python
import geopandas as gpd

read_shp_file(DATA_DIRECTORY / 'SourceDataFig2.shp')

# Leer el archivo .shp
df = gpd.read_file(DATA_DIRECTORY / 'SourceDataFig2.shp')

# Representar la columna 'geometry'
df['geometry'].plot(figsize=(20, 20))
```

```python
# Proyectar a un CRS proyectado
df_projected = df.to_crs('EPSG:3395')  # Mercator

# Calcular el centroide
df_projected['centroid'] = df_projected['geometry'].centroid

# Proyectar los centroides de vuelta a un CRS geográfico
df_projected['Lon'] = df_projected['centroid'].to_crs('EPSG:4326').x
df_projected['Lat'] = df_projected['centroid'].to_crs('EPSG:4326').y

df_projected_spain = df_projected.loc[(df_projected['Lat'] >= 35) & (df_projected['Lat'] <= 43) & (df_projected['Lon'] >= -10) & (df_projected['Lon'] <= 3)]
df_projected_spain
```
